```python
import matplotlib.pyplot as plt
import keras_ocr

pipeline = keras_ocr.pipeline.Pipeline()

images = [
    keras_ocr.tools.read(url) for url in [
        'https://upload.wikimedia.org/wikipedia/commons/b/bd/Army_Reserves_Recruitment_Banner_MOD_45156284.jpg', 
        '/content/drive/MyDrive/Colab Notebooks/printed1.PNG', 
        '/content/drive/MyDrive/Colab Notebooks/hw1.PNG', 
        '/content/drive/MyDrive/Colab Notebooks/hw2.PNG'
    ]
]

prediction_groups = pipeline.recognize(images)

fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

# easyocr

```python
# !pip install easyocr

import easyocr
reader = easyocr.Reader(['en'])
output = reader.readtext('/content/drive/MyDrive/Colab Notebooks/second2.jpg', detail = 0)
output1 = reader.readtext('/content/drive/MyDrive/Colab Notebooks/second2.jpg', detail = 0, paragraph=True)
print(output)
print("__________________________________")
print(output1)
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/second2.jpg')
res = reader.readtext('/content/drive/MyDrive/Colab Notebooks/second2.jpg') 
for (bbox, text, prob) in res: 
  # unpack the bounding box
  (tl, tr, br, bl) = bbox
  tl = (int(tl[0]), int(tl[1]))
  tr = (int(tr[0]), int(tr[1]))
  br = (int(br[0]), int(br[1]))
  bl = (int(bl[0]), int(bl[1]))
  cv2.rectangle(image, tl, br, (0, 255, 0), 2)
  cv2.putText(image, text, (tl[0], tl[1] - 10),
    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
plt.rcParams['figure.figsize'] = (16,16)
plt.imshow(image)
```
